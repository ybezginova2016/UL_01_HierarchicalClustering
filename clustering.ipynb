```python
def build_dendrogram(*args, **kwargs):
    max_d = kwargs.pop('max_d', None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)
    
    ddata = dendrogram(*args, **kwargs)
    
    if not kwargs.get('no plot', False):
        plt.title('Hierarchical Clustering Dendrogram(truncated)')
        plt.xlabel('Cluster size')
        plt.ylabel('Distance')
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y > annotate_above:
                plt.plot(x, y, 'o', c * c)
                plt.annotate('%.3g' % y, (x, y), xytest = (0, -5), textcoords = 'offset points', va = 'top', ha = 'center')
                
            if max_d:
                plt.axhline(y=max_d, c = 'k')
        return ddata
```

## 2.3 Cluster Evaluation

```python
from sklearn.metrics import v_measure_score
from sklearn.metrics.cluster import homogeneity_score
from sklearn.metrics.cluster import completeness_score
```

### Homogeneity metric
Homogeneity metric of a cluster labeling given a ground truth.

A clustering result satisfies homogeneity if all of its clusters contain only data points which are members of a single class.

This metric is independent of the absolute values of the labels: a permutation of the class or cluster label values won’t change the score value in any way.

```python
print('Homogeneity Score for different number of clusters:')
print('Number of clusters is two:', homogeneity_score(y_1d, pred_2))
print('Number of clusters is five:', homogeneity_score(y_1d, pred_5))
print('Number of clusters is seven:', homogeneity_score(y_1d, pred_7))
print('Number of clusters is ten:', homogeneity_score(y_1d, pred_10))
print('Number of clusters is fifteen:', homogeneity_score(y_1d, pred_15))
print('Number of clusters is thrity:', homogeneity_score(y_1d, pred_30))
```

### Completeness metric
Compute completeness metric of a cluster labeling given a ground truth.

A clustering result satisfies completeness if all the data points that are members of a given class are elements of the same cluster.

This metric is independent of the absolute values of the labels: a permutation of the class or cluster label values won’t change the score value in any way.

```python
print('Homogeneity Score for different number of clusters:')

print('Number of clusters is two:', completeness_score(y_1d, pred_2))
print('Number of clusters is five:', completeness_score(y_1d, pred_5))
print('Number of clusters is seven:', completeness_score(y_1d, pred_7))
print('Number of clusters is ten:', completeness_score(y_1d, pred_10))
print('Number of clusters is fifteen:', completeness_score(y_1d, pred_15))
print('Number of clusters is thrity:', completeness_score(y_1d, pred_30))
```

### V-Measure of cluster quality

```python
print('V-Measure for different number of clusters:')

print('Number of clusters is two:', v_measure_score(y_1d, pred_2))
print('Number of clusters is five:', v_measure_score(y_1d, pred_5))
print('Number of clusters is seven:', v_measure_score(y_1d, pred_7))
print('Number of clusters is ten:', v_measure_score(y_1d, pred_10))
print('Number of clusters is fifteen:', v_measure_score(y_1d, pred_15))
print('Number of clusters is thirty:', v_measure_score(y_1d, pred_30))
```

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    
    for i, merge in enumerate(model.children_):
        current_count = 0
        
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs);
```

```python
plt.figure(figsize=(39, 19))
plt.plot(color='fuchsia', linewidth=4)

plot_dendrogram(agg_cluster, truncate_mode="level", p=5)

plt.title('Dendrogram, Agglomerative Clustering', fontsize=37)
plt.xticks(fontsize=18)
plt.yticks(fontsize=30)
plt.xlabel('Number of points in the node', fontsize=30)
plt.ylabel('Values', fontsize=30)
plt.show();
```
